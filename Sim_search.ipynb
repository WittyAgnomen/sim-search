### Getting Started
First we will import graphlab and load our data into an SFrame. For convenience and best silimar product search we will perform nearest neighbors on each catergory's extracted features seperately. We will start with women's apparel. Data sets are located <a href="https://github.com/WittyAgnomen/sim-search/blob/master/caffe_macy.py" target="_blank">here</a>.

```python
import graphlab as gl
```

```python
#Load women's data into SFrame
sf=gl.SFrame('./womenpf.gl/')
```

```python
sf=sf.dropna()
#Add an id to each row
sf=sf.add_row_number()
#for later merging later
sf1=sf[['id','n']]
#print head of 
sf.head(5)
```

As we can see the data has a large amount of features, blah, blah check 2 x100!!

```python
#leave just the features in an sframe
sf4model=sf
sf4model.remove_columns(['id','cat','n'])

#initialize the model with the features
model = gl.nearest_neighbors.create(sf4model, distance='cosine')
#model summary
model.summary()
```

some in between writing


```python
#4 nearest neighbors
knn = model.query(sf4model, k=4)
knn.head()
```

```python
#adding names of products to knn reusults
sf1=sf1.rename({'id':'query_label'})
results=knn.join(sf1, on='query_label', how='left')
results=results.rename({'n':'product'})
sf1=sf1.rename({'query_label':'reference_label'})
results=results.join(sf1, on='reference_label', how='left')
results=results.rename({'n':'similar_product'})
```

```python
#see an example of the results
results.filter_by(12678, 'query_label')
```
